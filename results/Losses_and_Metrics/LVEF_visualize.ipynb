```python
data = pd.DataFrame({
    'minAreaRec':{'MSE':0.005457,'RMSE':0.061952,'Relative_Error':0.08675},
    'extremepoint':{'MSE':0.006419,'RMSE':0.070371,'Relative_Error':0.084738},
    'ellipse':{'MSE':0.005207,'RMSE':0.054632,'Relative_Error':0.068715}})

data = data.T
```

```python
sns.set_theme(style="ticks", palette="ch:s=.25,rot=-.25",font_scale=1.4)
plt.subplots(1,3,figsize=(3*3,6))
for i,n in enumerate(['MSE','RMSE','Relative_Error']):
    plt.subplot(1,3,i+1)
    sns.barplot(data=data.reset_index(),x='index',y=n,**dict(edgecolor="black", linewidth=2))
    plt.xticks(rotation=45)
    plt.xlabel('')
plt.tight_layout()
plt.show()
```
